```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection & preprocessing

```python
raw_mail_data=pd.read_csv('mail_data.csv')
raw_mail_data.head()
```

```python
mail_data=raw_mail_data.where((pd.notnull(raw_mail_data)),''
                              )

mail_data.head(5)
```

```python
#Checking the number of rows and cols from dataframe
mail_data.shape
```

Label encoding


```python
mail_data.loc[mail_data['Category']=="spam","Category",]==0
mail_data.loc[mail_data['Category']=="ham","Category",]==1
```

Spam=0
ham=1


```python
#Separting the data as text and label
X=mail_data['Message']
Y=mail_data['Category']
```

```python
print(X)
print(Y)
```

Spilting the data into training data and test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape)
print(X_test.shape)
print(X_train.shape)
```

Feature Extraction


```python
#Transform the text data to feature data that can be used as input to the logistic regression
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
#Converting Y_train and Y_test as integer

```

```python
print(X_train)
print(X_train_features)
```

Logistic Regression


```python
model=LogisticRegression()
```

```python
#Training the logistic regression model with training data
model.fit(X_train_features,Y_train)
```

Evaluating the model


```python
#Prediction on training data
prediction_on_training_data=model.predict(X_train_features)
accuracy_score_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
```

```python
print(accuracy_score_on_training_data)
```

```python
#Prediction on testing data
prediction_on_testing_data=model.predict(X_test_features)
accuracy_score_on_testing_data=accuracy_score(Y_test,prediction_on_testing_data)
```

```python
print(accuracy_score_on_testing_data)
```

Building the predictive system

```python
input_mail=[" Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's"]
#convert text to features vector
input_mail_features=feature_extraction.transform(input_mail)
print(input_mail_features)
```

```python
#Making the prediction
prediction=model.predict(input_mail_features)
print(prediction)
```

```python
if prediction[0]==1:
    print("Harm mail")
else:
    print("Spam mail")
```
